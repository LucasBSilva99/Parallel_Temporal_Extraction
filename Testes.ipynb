```python
def get_distribution(df, column, c='tab:blue', plot=False):
    values = pd.value_counts(df[column]).sort_index()
    if plot:
        ax = values.plot.bar(figsize=(30, 5), color=c, title=column)
        for p in ax.patches:
            ax.annotate(p.get_height(), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points')
        plt.show()
    return values.to_frame().transpose()
```
